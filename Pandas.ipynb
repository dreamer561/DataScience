```python
#lacture 11
# merging
left=pd.DataFrame({
      'key':['ko','k1','k2','k3'],
      'A':['Ao','A1','A2','A3'],
      'B':['Bo','B1','B2','B3'],
      
})
right=pd.DataFrame({
      'key':['ko','k1','k2','k3'],
      'C':['Co','C1','C2','C3'],
      'D':['Do','D1','D2','D3'],
      
})
```

```python
import pandas as pd
left=pd.DataFrame({
      'key1':['k0','k1','k2','k3'],
      'key2':['k0','k1','k0','k1'],
      'A':['A0','A1','A2','A3'],
      'B':['B0','B1','B2','B3'],
      
})
right=pd.DataFrame({
      'key1':['k0','k1','k2','k3'],
      'key2':['k0','k0','k0','k0'],
      'C':['C0','C1','C2','C3'],
      'D':['D0','D1','D2','D3'],
      
})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,on=['key1','key2'])
```

```python
pd.merge(left,right,on=['key1','key2'],how='outer')
```

```python
pd.merge(left,right,on=['key1','key2'],how='right')
```

```python
pd.merge(left,right,on=['key1','key2'],how='left')
```
