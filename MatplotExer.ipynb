```python
import numpy as np
x=np.arange(0,100)
y=x*2
z=x**2
import matplotlib.pyplot as plt
%matplotlib inline
df=plt.plot(y,z)
```

```python
# Use add_axes to add an axis
#  to the figure canvas at [0,0,1,1]. 
#  Call this new axis ax.
# Plot (x,y) on that axes and set
#  the labels and titles to match the plot below:

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.set_xlabel('X level')
ax.set_ylabel('Y level')
ax.plot(x,y,'r')
```

```python
# Create a figure object and put two axes on it, 
# ax1 and ax2. Located at 
# [0,0,1,1] and [0.2,0.5,.2,.2]
#  respectively.
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.2,.2])

# ax1.plot()

# ax2.plot()
```

```python
fig=plt.figure()
ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,.2,.2])

ax1.plot(x,y,'r')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax2.plot(x,y,'g')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
```

```python
# Create the plot below by adding two axes
#  to a figure object at [0,0,1,1] and
#   [0.2,0.5,.4,.4]

fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes1=fig.add_axes([0.2,0.5,.4,.4])
```

```python
fig=plt.figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.5,.4,.4])

axes1.plot(x,z)
axes1.set_xlabel('X')
axes1.set_ylabel('Y')


axes2.plot(y,x)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Zoom')
```

```python
# Now plot (x,y) and (x,z) on the axes.
# Play around with the linewidth and style
fig=plt.subplots(nrows=1,ncols=2)
plt.subplot(1,2,1)
plt.plot(x,y,'b--',lw=3.0,)

plt.subplot(1,2,2)
plt.plot(x,z,'r')
```

```python
fig=plt.subplots(nrows=1,ncols=2,figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(x,y,'b--',lw=3.0,)

plt.subplot(1,2,2)
plt.plot(x,z,'r')
```
